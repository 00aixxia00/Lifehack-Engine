## Average Precision

* Calculate precision@k at ranks with relevant documents ("X")
* divide by total number of relevant documents for topic
* total number of relevant documents is max(num(relevant docs)) from both systems

```python
# calculate max number of relevant documents for every topic
relevant_docs = {}
for i in range(len(lse_miri.index)):
    lse_count = lse_miri.iloc[i:i+1,1:13].values.tolist()[0].count('X')
    le_count = le_miri.iloc[i:i+1,1:13].values.tolist()[0].count('X')
    topic = i
    relevant_docs[topic] = max(lse_count, le_count)
relevant_docs  
```

```python
range(len(lse_miri.index))
```

```python
lse_miri["Expected # of rel. docs"] = relevant_docs.values()
le_miri["Expected # of rel. docs"] = relevant_docs.values()
```

```python
lse_miri
```

```python
le_miri
```

```python
def average_precision(row):
    judgements = row.tolist()[1:11]
    num_rel_docs = row.tolist()[11]
    
    # calculate only if number of relevant docs is not zero
    if num_rel_docs != 0:
        # p_at_k stores precision@k of relevant documents
        p_at_k = []
        # n counts number of relevant docs during iteration
        n = 0
        for i in range(10):
            rank = i + 1
            if judgements[i] == 'X':
                n = n + 1
                p = n/rank
                p_at_k.append(p)
        average_precision = sum(p_at_k)/num_rel_docs
    else:
        average_precision = np.nan
    return average_precision
```

```python
ex_col = lse_miri.iloc[:,12]
ex_col
```

```python
def mean_average_precision(col):
    col = col.tolist()
    m_a_p = sum(col)/20
    return m_a_p
```

```python
mean_average_precision(ex_col)
```

```python
le_miri["Average Precision"] = le_miri.apply(average_precision, axis=1).round(2)
display(le_miri)
print("MAP: " + str(mean_average_precision(le_miri.iloc[:,12])))
```

```python
lse_miri["Average Precision"] = lse_miri.apply(average_precision, axis=1).round(2)
display(lse_miri)
print("MAP: " + str(mean_average_precision(lse_miri.iloc[:,12])))
```

```python
compare = pd.DataFrame()
compare = lse_miri[['Topic']]
compare['Expected # of rel. docs'] = le_miri['Expected # of rel. docs'].values
compare['Average Precision LE'] = le_miri['Average Precision'].values
compare['Average Precision LSE'] = lse_miri['Average Precision'].values
compare['Difference'] = compare['Average Precision LE'] - compare['Average Precision LSE']
compare
```
